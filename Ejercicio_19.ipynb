# Ejercicio 19: Ecuación de Primer Grado

## ► Enunciado:

Crear un Algoritmo que permita hallar la solución a una ecuación de primer grado, de la forma: **ax + b = 0**  

El objetivo es despejar la **x** y validar los posibles datos para arrojar la respuesta.  

Al despejar la **x** tendremos que:  

**x = -b/a**  

Por lo tanto tendremos los siguientes escenarios:  

1) Si **a** es **DIFERENTE** de 0 (a != 0) la ecuación tiene solución.  

2) Si **a** es **IGUAL** a 0 (a == 0) tendremos que:  

- Si **b** es **DIFERENTE** de 0 (b != 0) la ecuación **no tiene solución**.  

- Si **b** es IGUAL a 0 (b == 0) la ecuación tiene **Infinitas Soluciones**.  
___  

## ► Variables:  

- **a**: Coeficiente principal.  

- **b**: Término Independiente.  

- **x**: Incógnita.  
___  

## ► Datos de Prueba.  

**a)** a = 2 y b = 6.  

**b)** a = 0 y b = 3.  

**c)** a = 0 y b = 0.  
___  

## ► Salida:  

![Texto alternativo](ejercicio19.png)


## a) a = 2 y b = 3

```python
# Valores para dibujar la tabla
ANCHO = 60
RELLENO1 = '-'
RELLENO2 = ' '
CADENA_VACIA = ''

# Mensajes a mostrar
MENSAJE_INICIAL = 'ECUACIÓN DE PRIMER GRADO: ax + b = 0'

# Declaración de variables
a = 2
b = 6

valor_de_a = f'>>> Valor de a: {a}'
valor_de_b = f'>>> Valor de b: {b}'
ecuacion = f'ECUACIÓN: {float(a)} x + {float(b)} = 0'


if a != 0:
    x = -b/a
else:
    if b != 0:
        x = 'La ecuación no tiene solución'
    else:
        x = 'La ecuación tiene infinitas soluciones'

solucion = f'>>> SOLUCIÓN: x = {x}'

# LINEA 1: Parte superior de la Tabla
print(CADENA_VACIA.center(ANCHO,RELLENO1))
# Mensaje Centrado
print(MENSAJE_INICIAL.center(ANCHO,RELLENO2))
# LINEA 2: Debajo del mensaje inicial
print(CADENA_VACIA.center(ANCHO,RELLENO1))
# Valores de a y b
print(valor_de_a.ljust(ANCHO,RELLENO2))
print(valor_de_b.ljust(ANCHO,RELLENO2))
# LINEA 3: Debajo de los valores de variables
print(CADENA_VACIA.center(ANCHO,RELLENO1))
# Ecuación
print(ecuacion.ljust(ANCHO,RELLENO2))
# LINEA 4: Debajo de la ecuación
print(CADENA_VACIA.center(ANCHO,RELLENO1))
# Solución
print(solucion.ljust(ANCHO,RELLENO2))
# LINEA 5: Parte inferior de la Tabla
print(CADENA_VACIA.center(ANCHO,RELLENO1))
```
